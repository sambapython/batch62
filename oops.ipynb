```python
class Emp:
    def __init__(self,name,e_id):
        self.name=name
        self.id=e_id
    def __add__(self1,self2):
        return " %s is a mother of %s"%(self1.name,self2.name)
    def __sub__(self1,self2):
        return self1.name+self2.name
    def __iter__(self):
        yield {"name":self.name}
        yield {"id":self.id}

sai=Emp("sailaxmi","2")
jay=Emp("jayaram","3")
print(sai+jay)
print(sai-jay)
for i in sai:
    print(i)
```

```python
class Emp:
    def __init__(self,name,e_id):
        self.name=name
        self.id=e_id
    def __add__(self1,self2):
        return " %s is a mother of %s"%(self1.name,self2.name)
    def __sub__(self1,self2):
        return self1.name+self2.name
    def __iter__(self):
        for i in self.__dict__.items():
            yield i
sai=Emp("sailaxmi","2")
jay=Emp("jayaram","3")
print(sai+jay)
print(sai-jay)
for i in sai:
    print(i)
```
